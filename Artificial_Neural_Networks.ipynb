```python
#  artificial neural networks
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV,ShuffleSplit
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import scale
from sklearn import model_selection
from sklearn.tree import DecisionTreeRegressor ,DecisionTreeClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
hit = pd.read_csv("/content/drive/MyDrive/Ornek_calÄ±smalar/Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[["League","Division","NewLeague"]])
y = df["Salary"]
X_ = df.drop(["Salary","League","Division","NewLeague"],axis=1).astype("float64")
X = pd.concat([X_,dms[["League_N","Division_W","NewLeague_N"]]],axis =1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
mlp_model = MLPRegressor().fit(X_train_scaled,y_train)  # model kuruldu.
mlp_model
```

```python
mlp_model.n_layers_  #katman sayÄ±sÄ±na eriÅim.
```

```python
mlp_model.hidden_layer_sizes  #gizli katman sayÄ±sÄ±na eriÅim.100
```

```python
# TAHMÄ°N
mlp_model.predict(X_train_scaled)[0:5]  # 5 gÃ¶zlem tahmini
```

```python
y_pred = mlp_model.predict(X_test_scaled)
np.sqrt(mean_squared_error(y_test,y_pred))   #test hatasÄ±
```

```python
# MODEL TUNÄ°NG
mlp_model
```

```python
mlp_params = {'alpha' : [0.1,0.01,0.02,0.005],
              'hidden_layer_sizes' : [(20,20),(100,50,150),(300,200,150)],
              'activation' : ['relu','logistic']}
```

```python
from sklearn.model_selection import GridSearchCV

mlp_cv_model = GridSearchCV(mlp_model,mlp_params,cv=10)
```

```python
mlp_cv_model.fit(X_train_scaled,y_train)
```

```python
mlp_cv_model.best_params_
```

```python
mlp_tuned = MLPRegressor(alpha=0.005,hidden_layer_sizes=(100,50,150))  # tune edilmiÅ model.
```

```python
mlp_tuned.fit(X_train_scaled,y_train)
```

```python
y_pred = mlp_tuned.predict(X_test_scaled)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
