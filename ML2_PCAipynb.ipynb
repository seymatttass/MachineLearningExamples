```python
import pandas as pd
import numpy as np

hit = pd.read_csv("/content/drive/MyDrive/Ornek_calÄ±smalar/Hitters.csv")
df = hit.copy()
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
dms.head()
```

```python
dms.info()
```

```python
y = df["Salary"]
```

```python
X_ =  df.drop(["Salary","League","Division","NewLeague"], axis = 1).astype("float64")
X_.head()
```

```python
X = pd.concat([X_,dms[["League_N","Division_W","NewLeague_N"]]],axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split


X_train,x_test,y_train,y_test = train_test_split(X,
                                                 y,
                                                 test_size=0.25,
                                                 random_state = 42)
print("X_train",X_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)

training = df.copy()

print("training",training.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

pca = PCA()
X_reduced_train = pca.fit_transform(scale(X_train))     # boyut indirgedik.
```

```python
X_reduced_train[0:1,:]       # 1.gÃ¶zlemm tÃ¼m bileÅenler
```

```python
np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)[0:5]

# 1.bileÅen veri setinde bulunan varyansÄ±n(toplam deÄiÅkenliÄin yÃ¼zde 38 ini aÃ§Ä±klamÄ±Å.)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
pcr_model = lm.fit(X_reduced_train,y_train)
```

```python
pcr_model.intercept_
```

```python
pcr_model.coef_
```

```python
# TAHMÄ°N

y_pred = pcr_model.predict(X_reduced_train)
y_pred[0:5]
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_train,y_pred)) # eÄitim hatasÄ±
```

```python
df["Salary"].mean()
```

```python
from sklearn.metrics import r2_score

r2_score(y_train,y_pred)    # r kare
```

```python
pca2= PCA()
```

```python
X_reduced_test = pca2.fit_transform(scale(x_test))
```

```python
y_pred = pcr_model.predict(X_reduced_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))    # test hatasÄ±
```

```python
# Model TUNÄ°NG

lm = LinearRegression()
pcr_model = lm.fit(X_reduced_train[:,0:10],y_train)
y_pred = pcr_model.predict(X_reduced_test[:,0:10])
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn import model_selection
```

```python
cv_10 = model_selection.KFold(n_splits=10,shuffle=True,random_state=1)
```

```python
lm = LinearRegression()
```

```python
RMSE = []
```

```python
for i in np.arange(1,X_reduced_train.shape[1]+1):
  score = np.sqrt(-1*model_selection.cross_val_score(lm,
                                                     X_reduced_train[:,0:i],
                                                     y_train,
                                                     cv=cv_10,scoring="neg_mean_squared_error").mean())
  RMSE.append(score)
```

```python
import matplotlib.pyplot as plt

plt.plot(RMSE,'-v')
plt.xlabel("BileÅen sayÄ±sÄ±")
plt.ylabel("RMSE")
plt.title("MaaÅ tahmin modeli iÃ§in PCR Model Tuning")
plt.show()

# optimum bileÅen sayÄ±sÄ± iÃ§in cross validation iÅlemi yaptÄ±k.
```
